```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/Mall_Customers.csv')
```

```python
print("Missing values per column:")
print(df.isnull().sum())
```

```python
df=df.dropna()
df=df.drop_duplicates()
```

```python
print(df)
```

```python
text_cols=['Gender']
for col in text_cols:
  if col in df.columns:
    df[col]=df[col].str.lower().str.strip()
```

```python
df['Gender']=df['Gender'].replace({
    'm':'Male',
    'f':'Female'
})
print(df.head)
```
